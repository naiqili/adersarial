```python
%matplotlib inline
from PIL import Image
from sklearn.cluster import MiniBatchKMeans, KMeans
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("test.jpg")
npimg = np.array(img)
height, width, channel = npimg.shape

print(width)
print(npimg.shape)
plt.figure(figsize=(16,16))
plt.imshow(npimg)
plt.show()
```

```python
pix = np.reshape(npimg, (-1, 3))

def kmeanlize(n_c):
    print('clusters: {}'.format(n_c))
    kmeans = KMeans(n_clusters=n_c).fit(pix)
    tmp_pix = np.zeros_like(pix)
    for idx, p in enumerate(pix):
        tmp_pix[idx] = kmeans.cluster_centers_[kmeans.labels_[idx]]

    tmp_pix = tmp_pix.reshape(npimg.shape)

    plt.figure(figsize=(16,16))
    plt.imshow(tmp_pix)
    plt.show()
```

```python
kmeanlize(2)
kmeanlize(5)
```
